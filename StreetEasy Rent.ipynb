```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


manhattan_df = pd.read_csv('https://raw.githubusercontent.com/caecario/datasets/master/streeteasy/manhattan.csv')
queens_df = pd.read_csv('https://raw.githubusercontent.com/caecario/datasets/master/streeteasy/queens.csv')
brooklyn_df = pd.read_csv('https://raw.githubusercontent.com/caecario/datasets/master/streeteasy/brooklyn.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(manhattan_df)

x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]

y = df[['rent']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

mlr = LinearRegression()

model=mlr.fit(x_train, y_train)

y_predict = mlr.predict(x_test)

# Input code here:
print("Train score:")
print(mlr.score(x_train, y_train))
print(mlr.score(x_test,y_test))

residuals = y_predict - y_test
```

```python
plt.scatter(y_predict, residuals, alpha=0.4)
plt.title('Residual Analysis')

plt.show()
```
